```python
#understanding composition of color image
def show_color_channel(image):
    color = ('r','g','b')
    for i,col in enumerate(color):
        channel=image[:,:,i]
        plt.subplot(1,3,i+1)
        plt.imshow(channel,cmap='gray')
        plt.title(f'{col} channel')
        plt.axis('off')
    plt.show()
show_color_channel(c)
```
